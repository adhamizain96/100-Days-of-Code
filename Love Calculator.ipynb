```python
person_1 = input('What is your name? \n')
person_2 = input('What is their name? \n')

couple = person_1 + person_2

t = couple.lower().count('t')
r = couple.lower().count('r')
u = couple.lower().count('u')
e = couple.lower().count('e')
true = t + r + u + e

l = couple.lower().count('l')
o = couple.lower().count('o')
v = couple.lower().count('v')
e = couple.lower().count('e')
love = l + o + v + e

compatability_score = true + love

if (compatability_score < 30) or (compatability_score > 60):
  print(f"Your score is {compatability_score}. You guys are very compatible with one another.")
elif (compatability_score >= 40) and (compatability_score <= 50):
  print(f"Your score is {compatability_score}. You should start seeing other people.")
else:
  print(f"Your score is {compatability_score}. You should probably break up with one another.")
```
