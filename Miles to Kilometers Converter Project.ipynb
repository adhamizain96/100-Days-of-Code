```python
#tkinter - standard Python interface to the Tcl/Tk GUI toolkit
from tkinter import *
def miles_to_km():
    #dictionary.get(keyname, value)
    #get() - returns the value of the item with the specified key
    miles = float(miles_input.get())
    km = round(miles * 1.609344, 3)
    km_result_ans.config(text = f'{km}')
#Tk() - allows you to register and unregister a callback function which will be called from the Tk mainloop
window = Tk()
window.title('Miles to Kilometers Converter Calculator')
#padx and pady - a distance - designating external padding on each side of the slave widget
window.config(padx = 25, pady = 25)

miles_input = Entry(width = 7)
miles_input.grid(column = 1, row = 0)
miles_label = Label(text = 'Miles')
miles_label.grid(column = 2, row = 0)
is_equal_label = Label(text = 'is equal to')
is_equal_label.grid(column = 0, row = 1)
km_result_ans = Label(text = '0')
km_result_ans.grid(column = 1, row = 1) 
km_result_label = Label(text = 'Kilometers')
km_result_label.grid(column = 2, row = 1)
#command calls the function
calculate = Button(text = 'Calculate', command = miles_to_km)
calculate.grid(column = 1, row = 2)

#mainloop() - tells Python to run the Tkinter event loop
window.mainloop()
```
