```python
import turtle as t
import random

nifty = t.Turtle()
nifty.speed('fastest')
# pensize() - Set the line thickness to width or return it
nifty.pensize(15)

directions = [0, 90, 180, 270]
color_list = ['CornflowerBlue', 'DarkOrchid', 'IndianRed','DeepSkyBlue',
              'LightSeaGreen', 'wheat', 'SlateGray', 'SeaGreen']

for _ in range(200):
    nifty.color(random.choice(color_list))
    nifty.forward(30)
    nifty.setheading(random.choice(directions))

nifty_image = t.Screen()
nifty_image.exitonclick()
```

```python
import turtle as t
import random

spiral_pic = t.Turtle()
t.colormode(255)

def random_color():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    color = (r, g, b)
    return color

spiral_pic.speed('fastest')

def draw_spirograph(gap_size):
    for _ in range(int(360/gap_size)):
        spiral_pic.color(random_color())
        spiral_pic.circle(100)
        #.setheading() - set the orientation of the turtle to to_angle.
        spiral_pic.setheading(spiral_pic.heading() + gap_size)

draw_spirograph(3)

spiral_pic_image = t.Screen()
spiral_pic_image.exitonclick()
```
