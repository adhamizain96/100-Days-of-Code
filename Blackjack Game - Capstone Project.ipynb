```python
import random

def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    card = random.choice(cards)
    return card

def calc_score(cards):
    if sum(cards) == 21 and len(cards) == 2:
        return 0
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(user_score, computer_score):
    if user_score > 21 and computer_score > 21:
        return 'You lose.'
    if user_score == computer_score:
        return 'It is a draw!'
    elif computer_score == 0:
        return 'You win!'
    elif user_score == 0:
        return 'The computer won!'
    elif user_score > 21:
        return 'You lose!'
    elif computer_score > 21:
        return 'You win!'
    elif user_score > computer_score:
        return 'You win!'
    else:
        return 'You lose!'

def play_game():
    user_cards = []
    computer_cards = []
    is_game_over = False
    for _ in range(2):
        new_card = deal_card()
        user_cards.append(new_card)
        computer_cards.append(new_card)
    while not is_game_over:
        user_score = calc_score(user_cards)
        computer_score = calc_score(computer_cards)
        print(f"Your cards are: {user_cards}, and your score is: {user_score}")
        print(f"The computer's first card is: {computer_cards[0]}")
        if user_score == 0 or computer_score == 0 or user_score > 21:
            is_game_over = True
        else:
            user_should_deal = input("Type 'Yes' to get another card, and type 'No' to pass: ")
            if user_should_deal == 'Yes':
                user_cards.append(deal_card())
            else:
                is_game_over = True    
        while computer_score != 0 and computer_score < 17:
            computer_cards.append(deal_card())
            computer_score = calc_score(computer_cards)
            print(f"Your final hand is: {user_cards}, and your final score is: {user_score}")
            print(f"Your computer's final hand is: {computer_cards}, and the computer's final score is: {computer_score}")
            print(compare(user_score, computer_score))
```

```python
play_game()
```
