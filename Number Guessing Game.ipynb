```python
from random import randint

Easy_Level_Turns = 10
Hard_Level_Turns = 5

def check_answer(guess, answer, turns):
    if guess > answer:
        print("The number you've guessed is too high!")
        return turns - 1
    elif guess < answer:
        print("The number you've guessed is too low!")
        return turns - 1
    else:
        print("You've guessed the correct number. Good Job!")
        
def set_difficulty():
    difficulty_level = input("Choose a difficulty level: Type 'Easy' or 'Hard': ")
    if difficulty_level == 'Easy':
        return Easy_Level_Turns
    else:
        return Hard_Level_Turns

def game():
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")
    answer = randint(1,100)
    turns = set_difficulty()
    guess = 0
    while guess != answer:
        print(f"You have {turns} turns left.")
        guess = int(input("Pick a number between 1 and 100: "))
        turns = check_answer(guess, answer, turns)
        if turns == 0:
            print(f"You have no more {turns} turns left. Try Again!")
            return
```

```python
game()
```
