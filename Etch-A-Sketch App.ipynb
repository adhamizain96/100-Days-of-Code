```python
import turtle as t

eas_app = t.Turtle()
eas_app_screen = t.Screen()

def move_forward():
    eas_app.forward(10)
def move_backward():
    eas_app.backward(10)
def turn_left():
    eas_app.left(10)
def turn_right():
    eas_app.right(10)
def clear_screen():
    eas_app.clear()
    # t.penup() - pull the pen up - no drawing when moving.
    eas_app.penup()
    #t.home() - move turtle to the origin â€“ coordinates (0,0)
    eas_app.home()
    eas_app.pendown()

eas_app_screen.listen()
eas_app_screen.onkey(move_forward, 'w')
eas_app_screen.onkey(move_backward, 's')
eas_app_screen.onkey(turn_left, 'a')
eas_app_screen.onkey(turn_right, 'd')
eas_app_screen.onkey(clear_screen, 'c')
eas_app_screen.exitonclick()
```
