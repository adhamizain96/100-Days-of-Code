```python
import turtle as t
import random

is_race_on = False
screen = t.Screen()
#screen.setup() - set the size and position of the main window.
screen.setup(width = 500, height = 500)
#x_pos
y_pos = [-70, -40, -10, 20, 50, 80]
#screen.textinput () - these popup input dialogs and return strings and numbers respectively.
race_bet = screen.textinput(title = 'Make your bet', prompt = 'Which turtle will win the race? Enter a color: ')
turtle_colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
turtles = []

for turtle_start in range(0, 6):
    new_turtle = t.Turtle(shape = 'turtle')
    #new_turtle.speed('fastest')
    #t.penup() - pull the pen up - no drawing when moving.
    new_turtle.penup()
    new_turtle.color(turtle_colors[turtle_start])
    #t.goto() - this method is used to move the turtle to an absolute position
    new_turtle.goto(x = -230, y = y_pos[turtle_start])
    turtles.append(new_turtle)
if race_bet:
    is_race_on = True
while is_race_on:
    for turtle in turtles:
        if turtle.xcor() > 230:
            is_race_on = False
            #t.pencolor() - return the current pencolor as color specification string or as a tuple
            winning_turtle = turtle.pencolor()
            if winning_turtle == race_bet:
                print('You have guessed correctly. Congrats!')
            else:
                print('You have guesssed incorrectly. Try Again!')
        turtle.forward(random.randint(0, 10))

screen.exitonclick()
```
