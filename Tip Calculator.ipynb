```python
bill = float(input("What was the total bill for tonight? $"))
tip = int(input("How much % tip would we all like to give? "))
people = int(input("How many people do we have to split the bill? "))

tip = bill * tip/100
total = bill + tip
split = total/people
split_amount = round(split, 2)
print(f"Each person pays ${split_amount}")
```
