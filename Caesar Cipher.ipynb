```python
letters = ['a','b','c','d','e','f','g',
           'h','i','j','k','l','m','n',
           'o','p','q','r','s','t','u',
           'v','w','x','y','z']

def caesar(start_text, shift_amount, cipher_direction):
    end_text = ""
    if cipher_direction == 'decode':
        shift_amount *= -1
    for char in start_text:
        if char in letters:
            position = letters.index(char) + shift_amount
            end_text += letters[position]
        else:
            end_text += char
    print(f"This is the final text: {end_text}")
            

cipher_direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
start_text = input("Type your message:\n").lower()
shift_amount = int(input("Type the shift number:\n"))
caesar(start_text, shift_amount, cipher_direction)
```
