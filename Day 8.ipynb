```python
import math
```

```python
def paint_area_calc(height, width, cover):
    paint_cans = (height * width)/cover
    round_up_paint_cans = math.ceil(paint_cans)
    print(f"You'll need {round_up_paint_cans} cans of paint. Time to get started and get the work done!")

height = int(input("What's the height of the wall?: "))
width = int(input("What's the width of the wall?: "))
cover = int(input("What's the cover of the wall?: "))
paint_area_calc(height, width, cover)
```

```python
def prime_checker(num):
    is_prime = True
    for n in range(2, num):
        if num % n == 0:
            is_prime = False
    if is_prime:
        print("This is a prime number.")
    else:
        print("This is not a prime number.")

num = int(input("Check to see if this number is a prime number: "))
prime_checker(num)
```
